```python
# Sales Prediction using Linear Regression
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dummy data
data = {'Sales': [100, 200, 300, 400], 'Profit': [10, 20, 30, 40]}
df = pd.DataFrame(data)

X = df[['Sales']]
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

print("Prediction for 500 sales:", model.predict([[500]]))
```
